```python
# ===============================
# 📊 Fertilizer Dataset — Full EDA
# ===============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look professional
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# ===============================
# 1️⃣ Load Dataset
# ===============================

df = pd.read_csv("Fertilizer_Dataset.csv")

# Drop unwanted columns
df = df.drop(columns=["District_Name", "Link"], errors="ignore")

print("✅ Dataset Loaded Successfully\n")


# ===============================
# 2️⃣ Basic Information
# ===============================

print("🔹 First 5 Rows")
print(df.head(), "\n")

print("🔹 Dataset Shape")
print(df.shape, "\n")

print("🔹 Data Types")
print(df.dtypes, "\n")

print("🔹 Info")
print(df.info(), "\n")


# ===============================
# 3️⃣ Missing Values Analysis
# ===============================

print("🔹 Missing Values")
missing = df.isnull().sum()
print(missing)

# Visualize missing values
plt.figure()
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()


# ===============================
# 4️⃣ Statistical Summary
# ===============================

print("🔹 Numerical Summary")
print(df.describe())

print("\n🔹 Categorical Summary")
print(df.describe(include="object"))


# ===============================
# 5️⃣ Unique Values in Each Column
# ===============================

print("\n🔹 Unique Values Per Column")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")


# ===============================
# 6️⃣ Categorical Analysis
# ===============================

categorical_cols = df.select_dtypes(include=["object"]).columns

for col in categorical_cols:
    print(f"\n🔹 Value Counts for {col}")
    print(df[col].value_counts())

    plt.figure()
    sns.countplot(data=df, x=col)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()


# ===============================
# 7️⃣ Numerical Distribution Analysis
# ===============================

numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    
    plt.show()


# ===============================
# 8️⃣ Boxplots for Outlier Detection
# ===============================

for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()


# ===============================
# 9️⃣ Correlation Analysis
# ===============================

corr = df.corr(numeric_only=True)

print("\n🔹 Correlation Matrix")
print(corr)

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()


# ===============================
# 🔟 Pairplot (Feature Relationships)
# ===============================

sns.pairplot(df[numerical_cols])
plt.show()


# ===============================
# 1️⃣1️⃣ Outlier Detection (IQR Method)
# ===============================

print("\n🔹 Outlier Detection using IQR")

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")


# ===============================
# 1️⃣2️⃣ Feature Correlation with Target (if exists)
# ===============================

# Replace 'Fertilizer' with your actual target column if needed
target_col = "Fertilizer"

if target_col in df.columns:
    print(f"\n🔹 Relationship with Target: {target_col}")

    for col in numerical_cols:
        if col != target_col:
            plt.figure()
            sns.scatterplot(x=df[col], y=df[target_col])
            plt.title(f"{col} vs {target_col}")
            plt.show()


print("\n✅ EDA Completed Successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Fertilizer_Dataset.csv")

# Ignore unwanted columns
df = df.drop(columns=["District_Name", "Link"], errors="ignore")


# =====================================================
# 🌿 1. NPK Distribution in ONE Attractive Graph
# =====================================================

plt.figure(figsize=(12, 7))

# Convert NPK columns into single format
npk_df = df[['Nitrogen', 'Phosphorus', 'Potassium']].melt(
    var_name='Nutrient',
    value_name='Value'
)

sns.violinplot(
    data=npk_df,
    x='Nutrient',
    y='Value',
    palette=['#27ae60', '#2980b9', '#f39c12']
)

plt.title("🌱 Distribution of Soil Nutrients (N-P-K)", fontsize=16, fontweight='bold')
plt.xlabel("Nutrient", fontsize=13)
plt.ylabel("Value", fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()


# =====================================================
# 🥧 2. Crop Distribution Pie Chart
# =====================================================

crop_counts = df['Crop'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    crop_counts,
    labels=crop_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("pastel"),
    wedgeprops={'edgecolor': 'black'}
)

plt.title("🌾 Crop Distribution", fontsize=15, fontweight='bold')
plt.show()


# =====================================================
# 🧪 3. Fertilizer Distribution Pie Chart
# =====================================================

fert_counts = df['Fertilizer'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    fert_counts,
    labels=fert_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("Set2"),
    wedgeprops={'edgecolor': 'black'}
)

plt.title("🧪 Fertilizer Distribution", fontsize=15, fontweight='bold')
plt.show()
```

### FERTILIZER DISTRIBUTION

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("Fertilizer_Dataset.csv")

# Ignore unwanted columns
df = df.drop(columns=["District_Name", "Link"], errors="ignore")


# ===============================
# 🧪 Fertilizer Distribution (Equal Divisions)
# ===============================

fert_counts = df['Fertilizer'].value_counts()

labels = fert_counts.index
percentages = fert_counts.values / fert_counts.values.sum() * 100

# ⚡ Equal sizes for all slices
equal_sizes = np.ones(len(labels))

# Gradient colors like your image
colors = plt.cm.viridis(np.linspace(0, 1, len(labels)))

plt.figure(figsize=(12, 10))

wedges, texts = plt.pie(
    equal_sizes,
    labels=labels,
    startangle=90,
    colors=colors,
    labeldistance=1.02,
    wedgeprops=dict(width=0.40)
)

# Add percentage text manually inside slices
for i, w in enumerate(wedges):
    angle = (w.theta2 + w.theta1) / 2
    x = 0.75 * np.cos(np.deg2rad(angle))
    y = 0.75 * np.sin(np.deg2rad(angle))
    plt.text(x, y, f"{percentages[i]:.1f}%", ha='center', va='center', fontsize=13)


# Legend on right
plt.legend(
    wedges,
    labels,
    title="Fertilizers",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.tight_layout()
plt.show()
```

### NPK DISTRIBUTION

```python
# ===============================
# 🌱 Grouped Bar Chart for NPK
# ===============================

import numpy as np
import matplotlib.pyplot as plt

# Calculate statistics
stats = {
    "Nitrogen": [
        df["Nitrogen"].mean(),
        df["Nitrogen"].median(),
        df["Nitrogen"].max()
    ],
    "Phosphorus": [
        df["Phosphorus"].mean(),
        df["Phosphorus"].median(),
        df["Phosphorus"].max()
    ],
    "Potassium": [
        df["Potassium"].mean(),
        df["Potassium"].median(),
        df["Potassium"].max()
    ]
}

labels = ["Mean", "Median", "Maximum"]
nutrients = list(stats.keys())

x = np.arange(len(nutrients))
width = 0.25

plt.figure(figsize=(10, 6))

# Bars for each statistic
plt.bar(x - width, [stats[n][0] for n in nutrients],
        width, label="Mean")

plt.bar(x, [stats[n][1] for n in nutrients],
        width, label="Median")

plt.bar(x + width, [stats[n][2] for n in nutrients],
        width, label="Maximum")

plt.xticks(x, nutrients)
plt.xlabel("Nutrients")
plt.ylabel("Value")

plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```
